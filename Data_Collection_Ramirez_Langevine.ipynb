Get News data about chatgpt, from the top 50 sites

```python
import pandas as pd
import requests
import time

results = pd.DataFrame([])
for domain in domains_50:
  # request data on google stock prices from thenewsapi.com
  url = "https://newsapi.org/v2/everything?q=ChatGPT&domains=" + domain + "&apiKey=b98e6cfa88d744ddad92a4b62da6a7ff"
  response = requests.request("GET", url )
  response_dict = response.json()
  print(response_dict.keys())
  # Extract the news data for chatGPT from the response dictionary
  news_data = response_dict['articles']

  # Convert the news data to a Pandas DataFrame
  news_df = pd.DataFrame(news_data)
  #news_df["Journal"] = domain
  #news_df["Ranking"] = domains_50[domain]
  results = pd.concat([results, news_df], ignore_index=True)
  
  time.sleep(0.5)
results
```

```python
results.to_csv("Feb19-Mar18.csv(1)", index=False) 
```
